```python
from pathlib import Path
import pandas as pd

# Load cleaned, merged dataset
df = pd.read_csv("data/cleaned_data.csv")

# Output directory
outdir = Path("data/af_inputs_fasta")
outdir.mkdir(parents=True, exist_ok=True)

# Determine sample ID column
id_col = "Sample_clean" if "Sample_clean" in df.columns else "Sample"

n = 0
for _, r in df.iterrows():
    sid = str(r[id_col])
    hc = str(r["VDJ_AA_H"])
    lc = str(r["VJ_AA_K"])

    fasta = f""">{sid}_H
{hc}
>{sid}_L
{lc}
"""

    with open(outdir / f"{sid}.fasta", "w") as f:
        f.write(fasta)
    n += 1

print(f"Wrote {n} paired FASTA files to {outdir}")
```
